```python
#Exercise 1.1
import numpy as np
def sample_mean(n):
    
    mat_random = np.random.normal(size=(n,n))
    sam_mean = np.mean(mat_random, axis=1)
    sam_var = np.var(sam_mean)
    
    return sam_var

def array_result(m,n):
    array_re = np.zeros((m,1))
    for i in range(m):
        array_re[i] = sample_mean(n) 
                        
    return array_re

plot_vec = array_result(100,100)
from matplotlib import pyplot as plt
plt.plot(plot_vec)
```
